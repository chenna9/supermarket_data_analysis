```python
categorical_values = ['Branch', 'City', 'Customer type', 'Gender', 'Product line', 'Payment']
for col in categorical_values:
    print(f"{col} unique values:\n{df[col].value_counts()}\n")
```

```python
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
import seaborn as sns
sns.countplot(data = df , x = 'Gender',palette=["#98d6f8", "#bff793"])
plt.title("Gender Distribution")
plt.show()
```

```python
import warnings
warnings.filterwarnings("ignore")
sns.countplot(data = df , x = 'Customer type',palette=["#98d6f8", "#bff793", "#f9c29f"])
plt.title("Customer Type")
plt.show()
```

```python



sns.countplot(data = df , x = 'Payment',palette=["#98d6f8", "#bff793", "#f9c29f"])
plt.title("Payment")
plt.show()
```

```python
sns.countplot(data = df , x = 'Branch')
plt.title("Number of Sales per Branch")
plt.show()
```

```python
branch_sales = df.groupby('Branch')['Sales'].sum().sort_values(ascending = False)
branch_sales
```

```python
branch_sales.plot(kind = 'bar' ,title = "Total sales per Branch" , color = 'green')
plt.ylabel("Total Sales")
plt.show()
                  
```

```python
city_transactions = df['City'].value_counts()
city_transactions.plot(kind = 'bar',title = 'Transactions per city')
plt.ylabel('no of Transactions')
plt.show()
```

```python
product_revenue = df.groupby('Product line')['Sales'].sum().sort_values(ascending = False)
product_revenue.plot(kind = 'bar' , title = 'Revenue per product line')
plt.show()
```
